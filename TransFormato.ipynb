# Transformación de datos de formato CSV a formato JSON

<p> A partir de los datos capturados en la actividad del tema 1, tendrás una hoja de trabajo que podrás descargar en formato CSV. Convierte los primeros cinco registros de ese fichero a un documento en formato JSON. Deberás proporcionar el fichero resultante con el documento JSON. </p>

## Cargar Librerias

```python
%matplotlib inline

import os
import pandas as pd
import numpy as np
```

## Cargar Dataset

```python
import codecs
ts = pd.read_csv('temperaturas_cleaned.csv', sep = ',')
ts = ts.drop('Unnamed: 0',axis=1)
ts.head(10)
```

## Convertir en JSON, los 5 primeros registros

```python
five = ts.iloc[0:5,]
cols = ['Latitud','Longitud','Enero', 'Febrero','Marzo','Abril','Mayo','Junio','Julio',
                      'Agosto','Septiembre','Octubre','Noviembre','Diciembre','Media', 'Media Real', 'Diferencia']
five = pd.DataFrame(five, columns= cols)
five.head()
```

```python
five.to_json('temperaturas_table.json',orient="table")
five.to_json('temperaturas_split.json',orient="split")
five.to_json('temperaturas_records.json',orient="records")
five.to_json('temperaturas_index.json',orient="index")
five.to_json('temperaturas_values.json',orient="values")
five.to_json('temperaturas_columns.json',orient="columns")
```
