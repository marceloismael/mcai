# Saneado de Datos

<p> Se solicita reparar los datos de un dataset, disponible en el aula virtual. 
No importa el lenguaje utilizado para repararlo. 
Dentro de los requisitos de saneamiento se pide: </p>

<ul><li>Generar un csv separado por comas</li>
<li>Que el carácter decimal sea el ‘.’ (punto)</li>
<li>Eliminar los caracteres extraños de longitud y latitud</li>
<li>Ponerle el símbolo de temperatura a todas las medidas</li>
<li>Comprobar que la media es la media aritmética de las temperaturas de los meses del año. En el caso que no sea correcta indicarlo en una columna adicional con un número que indique la desviación.</li>
</ul>

## Cargar Librerias

```python
%matplotlib inline

import os
import pandas as pd
import numpy as np
```

## Generar CSV

```python
ts = pd.read_csv('temperaturas.txt', sep = '\s+')
ts.rename(columns={'Septimbre':'Septiembre'}, inplace=True)
ts.to_csv('temperaturas.csv', index=False)
ts.head(10)
```

## Cargar Dataset

```python
ts = pd.read_csv('temperaturas.csv', sep = ',')
ts.head(10)
```

## Que el caracter decimal sea el punto (.)

```python
ts = ts.apply(lambda x: x.str.replace(',','.'))
ts.head(10)
```

## Eliminación de Caracteres Extraños

```python
ts = ts.apply(lambda x: x.str.replace('º',''))
ts.head(10)
```

### Cambio de tipos de datos como verificación de ausencia de valores extraños 

```python
#cambio de tipos de datos de columnas 
ts= ts.apply(lambda col:pd.to_numeric(col, errors='coerce'))
#ts['Media'] = ts['Media'].astype(np.float64)
#ts.dtypes
ts.describe()
```

## Comprobación de Media Aritmética

```python

ts['Media Real'] = ts[['Enero', 'Febrero','Marzo','Abril','Mayo','Junio','Julio',
                      'Agosto','Septiembre','Octubre','Noviembre','Diciembre']].mean(axis=1)
#axis = 1 means 'USE COLUMNS'
ts = ts.round({'Media Real':1})
ts['Diferencia'] = ts['Media'] - ts['Media Real']
tsx = ts
ts.to_csv('temperaturas_instance.csv', index=False, float_format='%.3f')
ts.to_json('temperaturas_complete_records.json',orient="records")
ts.head()
```

## Colocación de símbolo de temperatura

```python
cols = ['Enero', 'Febrero','Marzo','Abril','Mayo','Junio','Julio',
                      'Agosto','Septiembre','Octubre','Noviembre','Diciembre','Media', 'Media Real', 'Diferencia']

tsx[cols] = ts[cols].round({'Diferencia':1}).astype(str) + "º"
tsx.head()
```

## Guardar el Archivo saneado

```python
tsx.to_csv('temperaturas_cleaned.csv', index=False)
```
