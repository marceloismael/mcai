# Prototipo de documento para el almacenado de data de sensores

<p> La idea de este prototipo es generar una clase con los atributos que los sensores capturan. Luego a través de la carga de un archivo de texto configurar instancias de la clase creada y serializarlas en formato JSON para la carga en la base de datos. </p>

### Cargar Librerias

```python
%matplotlib inline

import os
import pandas as pd
import numpy as np
```

### Definición de clase Location y la clase "Environment" (Env) 

```python
class Location:
    def __init__(lat, lon):
        self.latitude = lat
        self.longitude = lon
        
class Env:
    def __init__(sensorId, datetime, lat, lon, temperature, humidity, noise, light, windspeed):
        self.sensorId =sensorId
        self.datetime = datetime
        self.location = Location(lat,lon)
        self.temperature = temperature
        self.humidity = humidity
        self.noise = noise
        self.light = light
        self.windspeed = windspeed
        
```

### Cargar Dataset

```python

sd = pd.read_csv('sensors_data.csv', sep = ',')
sd.head(10)
```

### Verificación de datos numéricos

### Cambio de tipo string a fecha

```python
#import pendulum
import time
from datetime import datetime
#mmm yy yyyy hh:mm AM/PM
#sd['timestamp'] = sd['timestamp'].apply(lambda x: datetime.strptime(x,'%b %d %Y %H:%M %p'))

#ISO DATE yyyy-mm-ddThh:MM:ss
sd['timestamp'] = sd['timestamp'].apply(lambda x: datetime.strptime(x,'%Y-%m-%dT%H:%M:%S'))
sd.head()
sd.dtypes
```

### Crear archivo JSON

```python
sd.to_json('sensors_data.json',orient='records',
           lines=True)
```
